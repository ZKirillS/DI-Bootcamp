```python
# Load an image using PIL
image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
```

```python
#rotating the image X number of degrees
rotated_image1 = original_image.rotate(90)
plt.imshow(rotated_image1)
```

```python
#rotating the image X number of degrees
rotated_image2 = original_image.rotate(50)
plt.imshow(rotated_image2)
```

```python
#Flip an image horizontally and then vertically using mirror and flip functions from ImageOps

flipped_image = ImageOps.flip(original_image)
plt.imshow(flipped_image)

```

```python

mirrored_image = ImageOps.mirror(original_image)
plt.imshow(mirrored_image)
```

```python
#Zoom in on an image (scale by 1.2x) using .resize(...)

resized_image = original_image.resize((150,150))
plt.imshow(resized_image)
```

```python
import random
import matplotlib.image as mpimg


list_img = []
plt.figure(figsize=(20,20))
for i in range(5):
  file = random.choice(image_path)
  img=mpimg.imread(image_path)
  list_img.append(img)

for i, img in enumerate(list_img):
  ax=plt.subplot(3,5,i+1)
  ax.title.set_text(file)
  plt.imshow(img)
```

```python
# To visualize the rotated images :
rotated_list = []
plt.figure(figsize=(20,20))
for img in list_img:
  rotated_list.append(rotated_image1)
  rotated_list.append(rotated_image2)

for i, img in enumerate(rotated_list):
    ax=plt.subplot(3,5,i+1)
    ax.title.set_text(file)
    plt.imshow(img)
```
