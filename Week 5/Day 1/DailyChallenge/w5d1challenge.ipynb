```python
min_health = world_h["Healthy life expectancy"].min()
max_health = world_h["Healthy life expectancy"].max()
world_h["Healthy life expectancy_Normalized"] = ((world_h["Healthy life expectancy"] - min_health) / (max_health - min_health)) * 100
print(world_h['Healthy life expectancy_Normalized'])
```

```python
import matplotlib.pyplot as plt
region_avg = world_h.groupby('Region').agg({
    'GDP per capita': 'mean',
    'Healthy life expectancy_Normalized': 'mean'
}).reset_index()

region_avg_sorted = region_avg.sort_values(by='GDP per capita', ascending=True)

x = region_avg_sorted['Region']
y1 = region_avg_sorted['GDP per capita']
y2 = region_avg_sorted['Healthy life expectancy_Normalized']

fig, ax1 = plt.subplots(figsize =(8,5))

ax1.plot(x,y1,marker='o', color = "blue")
ax1.set_xlabel("Region")
ax1.set_ylabel("GDP per capita", color="blue")
ax1.grid(True, linestyle='--', alpha=0.9)
ax1.tick_params(axis='x', rotation=45)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.set_ylabel("Healthy life expectancy", color="red")
ax2.plot(x,y2,marker='o', color = "red")
ax2.grid(True, linestyle='--', alpha=0.9)
ax2.legend(loc='upper right')

fig.suptitle("Comparison of Healthy life expectancy and GDP per capita")
plt.tight_layout()
plt.show()
```

**The stronger the economy, the greater the per capita income, and it can be concluded that the quality and standard of living increases accordingly. This directly affects the expected average standard of living of the country's population. In this sample, we look at the average indicators for entire regions and see the direct dependence of these indicators. Although in some regions there are countries with much stronger and much weaker economic situation (outliers), which significantly affects the average indicator, which does not reflect the real situation in a particular country, as for example we see in the Middle East.**
