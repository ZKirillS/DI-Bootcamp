```python
import pandas as pd

df = pd.read_csv('Airplane_Crashes_and_Fatalities.csv', encoding='latin1')
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df_clean = df.drop(columns = ['Time','Operator','Flight #','Registration','cn/ln'])
```

```python
df_clean.columns = df_clean.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df_clean.duplicated().sum()
```

```python
df_clean.info()
```

```python
df_clean['location'].fillna('Somewhere', inplace=True)
```

```python
df_clean['route'].fillna('Not a route', inplace=True)
```

```python
df_clean['ac_type'].fillna(df_clean['ac_type'].mode()[0], inplace=True)
```

```python
df_clean.dropna(subset=['aboard'], inplace=True)
```

```python
df_clean['summary'].fillna('Crash', inplace=True)
```

```python
df_clean['ground'].fillna(df_clean['ground'].mean(), inplace=True)
```

```python
df_clean['fatalities_crew']=df_clean.apply(
    lambda row: row['fatalities']/4 if pd.isnull(row['fatalities_crew']) and pd.notnull(row['fatalities']) else row['fatalities_crew'],
    axis=1
) # calculating that 1/4 of fatalities was a crew and the 3/4 was passengers and filling missing values with this numbers
df_clean['fatalities_passangers']=df_clean.apply(
    lambda row: (3 * row['fatalities'])/4 if pd.isnull(row['fatalities_passangers']) and pd.notnull(row['fatalities']) else row['fatalities_passangers'],
    axis=1
) # same calculating for 3/4 for passengers
```

```python
df_clean['aboard_crew']=df_clean.apply(
    lambda row: row['aboard']/5 if pd.isnull(row['aboard_crew']) and pd.notnull(row['aboard_crew']) else row['aboard_crew'],
    axis=1
) # calculating that 1/5 of aboard was a crew and the 3/5 was passengers and filling missing values with this numbers
df_clean['aboard_passangers']=df_clean.apply(
    lambda row: (4 * row['aboard'])/5 if pd.isnull(row['aboard_passangers']) and pd.notnull(row['aboard_passangers']) else row['aboard_passangers'],
    axis=1
) # same calculating for 4/5 for passengers
```

```python
df_clean['aboard'] = df_clean['aboard'].astype(int)
df_clean['aboard_passangers'] = df_clean['aboard'].astype(int)
df_clean['aboard_crew'] = df_clean['aboard'].astype(int)
df_clean['fatalities'] = df_clean['fatalities'].astype(int)
df_clean['fatalities_passangers'] = df_clean['fatalities_passangers'].astype(int)
df_clean['fatalities_crew'] = df_clean['fatalities_crew'].astype(int)
df_clean['ground'] = df_clean['ground'].astype(int)
# converting all numerical columns into integers for better visualisation and calculating
```

```python
df_clean.info()
```

```python
df_clean['date'] = pd.to_datetime(df_clean['date'])
df_clean['year'] = df_clean['date'].dt.year # pulling out a years for better visualisation time series
```

```python
df_clean.head()
```

```python
total_fatalities = df_clean['fatalities'].sum()
crashes_with_fatalities = df_clean[df_clean['fatalities'] > 0]
total_survivors = df_clean['aboard'].sum() - total_fatalities
crashes_with_survivals = df_clean[df_clean['aboard'] > df_clean['fatalities']]
survival_rate = crashes_with_survivals.shape[0] / crashes_with_fatalities.shape[0] * 100
print(f"Survival rate: {survival_rate:.2f}%")
print(f'Crashes with fatalities: {crashes_with_fatalities.shape[0]} of 4980')
print(f'Total fatalities: {total_fatalities}')
print(f'Total survivors: {total_survivors}')
```

**The percentage of survivors in the crash is 36.6, which is a pretty good indicator. Considering that almost 99% of crashes involved casualties, this survival rate is quite high.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
# calculating % of survivors in crashes with fatalitites
survival_rate_data = [crashes_with_survivals.shape[0], crashes_with_fatalities.shape[0] - crashes_with_survivals.shape[0]]
labels = ['Crashes with Survivors','Crashes without Survivors']
plt.figure(figsize=(7,7))
plt.pie(survival_rate_data, labels=labels, autopct='%1.1f%%')
plt.title('Survival Rate')
plt.axis('equal')
plt.show()
```

here we can notice that most accidents ended in the death of the entire crew and all passengers

```python
total_data= [total_fatalities, total_survivors]
labels = ['Fatalities','Survived']
plt.figure(figsize=(10,6))
sns.barplot(x=labels, y=total_data, palette='viridis')
plt.xlabel('Total Fatalities and Survived')
plt.ylabel('Count')
plt.title('Total Fatalities and Survived')
plt.grid(True, linestyle = '--')
plt.show()
```

Here we clearly see the ratio of dead and survivors. More than 110 thousand people have died in disasters over the years. However, given that this is an indicator for a period of more than 100 years, it can be assumed that this indicator is influenced by progress in the field of aircraft construction and safety.

```python
amount_crashes_per_year = df_clean.groupby(df_clean['year'])['year'].count()
plt.figure(figsize=(16,10))
sns.barplot(x=amount_crashes_per_year.index, y=amount_crashes_per_year.values, width = 0.6)
plt.xlabel('Year')
plt.ylabel('Amount of Crashes')
plt.xticks(rotation=80)
plt.title('Amount of Crashes per Year')
plt.grid(True, linestyle = '--')
plt.show()
```

```python
# Distribution of fatalities
import numpy as np
from scipy import stats
fatalities = df_clean['fatalities']
mean_fatalities = stats.tmean(fatalities) # calculating mean, median and standart deviation
median_fatalities = np.median(fatalities)
std_fatalities = stats.tstd(fatalities)
print(f'Mean: {mean_fatalities}')
print(f'Median: {median_fatalities}')
print(f'Standard Deviation: {std_fatalities}')

plt.figure(figsize=(10,6))
sns.histplot(df_clean['fatalities'], bins = 100, kde=True, color='red')
plt.title('Distribution of Fatalities')
plt.xlabel('Fatalities')
plt.ylabel('Frequency')
plt.grid(True, linestyle = '--')
plt.show()
```

Here we see that most often crashes lead to mass fatalities, this is due to the expansion of the carrying capacity of aircraft. Since crashes are often due to critical malfunctions, landing the plane safely is incredibly difficult. the average number of deaths in a disaster is 22. We also see that there have been disasters with a huge number of fatalities (std 35, it suggests that some crashes had very few fatalities, while others had a very large number of fatalities). We can see a huge peak, which could represent specific historical event or huge aircraft crash.

```python
# Survival rate distribution
import numpy as np
from scipy import stats
df_clean['survival_rate'] = (df_clean['aboard'] - df_clean['fatalities']) / df_clean['aboard'] * 100
# calculating total survival rate (crew and passengers)
survival_rates = df_clean['survival_rate'].dropna() # dropping any missing values to avoid errors
mean_survivors = stats.tmean(survival_rates)
median_survivors = np.median(survival_rates)
std_survivors = stats.tstd(survival_rates)
print(f'Mean: {mean_survivors}')
print(f'Median: {median_survivors}')
print(f'Standard Deviation: {std_survivors}')

plt.figure(figsize=(10,6))
sns.histplot(df_clean['aboard'] - df_clean['fatalities'], bins = 100, kde=True, color='green')
plt.title('Distribution of Survivors')
plt.xlabel('Survivors')
plt.ylabel('Frequency')
plt.grid(True, linestyle = '--')
plt.show()
```

relative to the previous graph, here we see a very small number of survivors, however, with the exception of outliers, the graph should be smooth

```python
# Conduct a hypothesis test (e.g., comparing the average number of fatalities in different decades or regions).
df_clean['decade'] = df_clean['year'].apply(lambda x: (x // 10) * 10)
decades = df_clean.groupby('decade')['fatalities'].mean()
avg_fatalities = df_clean.groupby('decade')['fatalities'].mean()
fatalities_1940 = df_clean[df_clean['decade']==1940]['fatalities']
fatalities_1950 = df_clean[df_clean['decade']==1950]['fatalities']
# Null hypothesis - there is a big difference in fatalities between 1940 and 1950 decades
# Alternative hypothesis - there is no big difference in fatalities between 1940 and 1950 decades
t_statistic, p_value = stats.ttest_ind(fatalities_1940, fatalities_1950)
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
if p_value < 0.05:
    print("there is a big difference in fatalities between 1940 and 1950 decades.")
else:
    print("there is no big difference in fatalities between 1940 and 1950 decades.")
```

```python
fatalities_1990 = df_clean[df_clean['decade']==1990]['fatalities']
fatalities_2000 = df_clean[df_clean['decade']==2000]['fatalities']
# Null hypothesis - there is a big difference in fatalities between 1990 and 2000 decades
# Alternative hypothesis - there is no big difference in fatalities between 1990 and 2000 decades
t_statistic, p_value = stats.ttest_ind(fatalities_1990, fatalities_2000)
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
if p_value < 0.05:
    print("there is a big difference in fatalities between 1990 and 2000 decades.")
else:
    print("there is no big difference in fatalities between 1990 and 2000 decades.")
```

```python
# calculating crew survival rate % to see correlation
df_clean['crew_survival_rate'] = ((df_clean['aboard_crew'] -df_clean['fatalities_crew']) / df_clean['aboard']) * 100
numeric_columns = df_clean.select_dtypes(include=['number']) # selecting all numerical columns for correlation
correlation_matrix = numeric_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_clean, y='crew_survival_rate', x='survival_rate', alpha = 0.7,color='red')
plt.title('Relationship between Crew survival rate and total survival Rate')
plt.ylabel('Crew survival rate %')
plt.xlabel('Survival Rate %')
plt.grid(True, linestyle = '--')
plt.show()
```

The survivability of the crew affects the overall survival rate, because if the crew survived, then most likely the landing was successful and casualties (or a large number of casualties) were avoided.
