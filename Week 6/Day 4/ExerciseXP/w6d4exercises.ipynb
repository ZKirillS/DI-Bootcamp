```python
# Exercise 1 : Linear Regression Analysis
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression
house_sizes= [50, 70, 80, 100, 120]
house_prices= [150000, 200000, 210000, 250000, 280000]
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)
print(f"Slope: {slope}, Intercept: {intercept}")
new_house = 90
predicted_price = slope * new_house + intercept
print(f'New price: {predicted_price}')
plt.scatter(house_sizes, house_prices, color='blue', label='Data Points')
plt.plot(house_sizes, [slope * x + intercept for x in house_sizes], color='red', label='Regression Line')
plt.xlabel('House Size (m^2)')
plt.ylabel('House Price ($)')
plt.title('House Size vs. Price')
plt.legend()
plt.show()
```

slope is showing us how predicted price is going up or down, if size of house is changing for 1 m2. 1828 (slope) is a price for 1 m2 for every house in this data.

```python
# Exercise 2 : Understanding ANOVA

fertilizer_1=[5, 6, 7, 6, 5]
fertilizer_2=[7, 8, 7, 9, 8]
fertilizer_3=[4, 5, 4, 3, 4]
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print(f'F-value: {f_value}, p-value: {p_value}')
plt.boxplot([fertilizer_1, fertilizer_2, fertilizer_3], labels=['Fertilizer 1', 'Fertilizer 2', 'Fertilizer 3'])
plt.ylabel('Plant Growth (cm)')
plt.title('Plant Growth by Fertilizer')
plt.show()
```

f-value is 28.5 - its a big number that tell usthe groups are quite different and dont have the same avg value.
p-value is less then 0.05 means that differences are real and not just by chance. So fertilizers have different effects on plant growth.
If the p-value were greater than 0.05 - differences could be just by chance

```python
# Exercise 3 : Probability Distributions
import matplotlib.pyplot as plt
from scipy.stats import binom
p_heads = 0.5  # Probability of getting heads
n_flips = 10  # Total customers
k = 5  # Exactly 5 heads

prob_binomial = binom.pmf(k, n=n_flips, p=p_heads)
print(f"\nProbability of exactly 5 out of 10 flips getting heads: {prob_binomial:.2f}")

# Visualization of Binomial Distribution
x_binom = range(0, n_flips + 1)
binomial_pmf = binom.pmf(x_binom, n=n_flips, p=p_heads)
plt.figure(figsize=(8, 5))
plt.bar(x_binom, binomial_pmf, color='green', alpha=0.7)
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.title('Binomial Distribution')
plt.show()
```

```python
# Exercise 4 : Correlation Coefficients
# Exercise 5 : Linear Relationships
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
from scipy.stats import pearsonr
correlation, p_value = pearsonr(data['age'], data['income'])
print(f'Correlation: {correlation}, p-value: {p_value}')
```

Correlation is close to 1, whicih indicates a perfect positive linear relationship. The higher x - the bigger y (income) is. Both variables increase together.

```python
from scipy.stats import spearmanr
correlation, p_value = spearmanr(data['age'], data['income'])
print(f'Correlation: {correlation}, p-value: {p_value}')
plt.scatter(data['age'], data['income'], color='green')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs. Income')
plt.show()
```
