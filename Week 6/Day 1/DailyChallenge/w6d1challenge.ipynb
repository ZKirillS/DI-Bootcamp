# Daily Challenge: NumPy, Pandas, and Matplotlib Integration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

city = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix",
        "Philadelphia", "San Antonio", "San Diego", "Dallas", "San Jose"]
months = ["January", "February", "March", "April", "May", "June",
         "July", "August", "September", "October", "November", "December"]
temperature = np.random.uniform(-5,35, size =(len(city),len(months)))

df = pd.DataFrame(data=temperature, index=city, columns=months)
print(df)
```

```python
temp_avg = df.mean(axis=1)
temp_avg_sorted = temp_avg.sort_values(ascending=False)
print(temp_avg_sorted)
```

```python
temp_high = temp_avg.idxmax()
temp_low = temp_avg.idxmin()
print(temp_high)
print(temp_low)
```

```python
df['Temp AVG']= temp_avg
plt.figure(figsize=(10,8))
plt.plot(df.columns, df.loc[temp_high], color='red')
plt.plot(df.columns, df.loc[temp_low], color='blue')
plt.title('Monthly Temp Trends for Chicago and San Antonio', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.xticks(rotation=45)
plt.ylabel('Temperature (Â°C)', fontsize=12)
plt.grid(True, color = 'black', linestyle='--', linewidth =0.3, alpha=0.7)
plt.show()
```

in cities with the lowest and highest average temperatures there is no order in temperature changes, it jumps from high to low from month to month

```python
plt.figure(figsize=(12, 6))

for city in df.index.unique():
    plt.plot(df.columns, df.loc[city], label=city)

plt.title('Monthly Temp Trends for each City', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Temperature (Â°C)', fontsize=12)
plt.grid(True, color = 'black', linestyle='--', linewidth =0.3, alpha=0.7)
plt.legend(loc='lower left')

plt.show()
```

the lowest temperature was observed in Dallas in November (-4.97), the highest in Los Angeles in July (34.68). The highest average temperature is in San Antonio (11.18), the lowest average in Chicago (11.1).
